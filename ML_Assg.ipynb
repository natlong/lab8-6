```python
import pandas as pd
import numpy as np
# from pypdf import PdfReader
# from tabula import read_pdf
# from tabulate import tabulate
import camelot

file_path = "data/tosp_data.pdf"

data = camelot.read_pdf(file_path, pages='all', flavor='stream')  # 'stream' or 'lattice' based on table structure

print(len(data))

df = data[0].df
print(df)

#Model Packages
```

```python
df.to_csv("output.csv", index=False)

# Display other tables if needed
for i, table in enumerate(data):
    if (i>5):
      print(f"Table {i+1}")
      print(table.df)
```

```python
full_df = []

for i, table in enumerate(data):
  if (i>5 and i <110):
    full_df.append(table.df)

df2 = pd.concat(full_df, ignore_index=True)
print(df2)
```

```python
df2.to_csv("total_output.csv", index=False)
```

```python

df2.columns = ['Num','S/N', 'Code', 'Description', 'Table', 'Classification']

#Cleaning data

cleaned_rows = []
current_row = None

for _, row in df2.iterrows():
  if pd.notnull(row['S/N']):
    if current_row is not None:
      cleaned_rows.append(current_row)
    current_row = row.copy()
  elif current_row is not None:
    current_row['Description'] += " " + str(row['Description']).strip()


if current_row is not None:
  cleaned_rows.append(current_row)

cleaned_df = pd.DataFrame(cleaned_rows)



print(cleaned_df)
```
